```python
import numpy as np
import matplotlib.pyplot as plt

population = np.random.exponential(scale=2.0, size=100000)

plt.hist(population, bins=100, density=True, alpha=0.7, color="orange")
plt.title("Exponential distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
true_mean = np.mean(population)
print("Approximately true mean:", true_mean)

sample_sizes = np.arange(10, 5000, 5)  
sample_means = []

for n in sample_sizes:
    sample = np.random.choice(population, size=n, replace=False)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(10,5))
plt.plot(sample_sizes, sample_means, label="Sample mean", color="blue")
plt.hlines(true_mean, xmin=0, xmax=sample_sizes[-1], colors="red", linestyles="dashed", label="Real mean")
plt.xlabel("Sample size")
plt.ylabel("Avg value")
plt.title("Law of large numbers")
plt.legend()
plt.show()
```
