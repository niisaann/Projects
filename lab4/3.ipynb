1.What is the primary purpose of the Central Limit Theorem (CLT)?
The primary purpose of the CLT is to show that, regardless of the original distribution of the population, the distribution of the sample mean will approach a normal distribution as the sample size becomes large. This allows us to apply methods of inference that rely on normality, even when the underlying data is not normally distributed.
