```python
import numpy as np
import matplotlib.pyplot as plt

population = np.random.exponential(scale=2.0, size=100000)

plt.hist(population, bins=100, density=True, alpha=0.7, color="orange")
plt.title("Exponential distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
sample_size = 50
num_samples = 1000
means = []
for i in range(num_samples):
    sample = np.random.choice(population, size = sample_size , replace= True)
    means.append(np.mean(sample))
plt.hist(means, bins=100, density=True, alpha=0.7, color="blue")
plt.title("Distribution with sample size 50 ")
plt.xlabel("Avg value")
plt.ylabel("Density")
plt.show()
```

```python
sample_size = 200
means_large = []

for i in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    means_large.append(np.mean(sample))

plt.hist(means_large, bins=100, density=True, alpha=0.7, color="green")
plt.title("Distribution with sample size 200")
plt.xlabel("Avg value")
plt.ylabel("Density")
plt.show()
```
