```python
import numpy as np 
import matplotlib.pyplot as plt

Y = np.random.uniform(low=0, high=10, size=100)

c_values = np.linspace(0, 10, 10)  
```

4.2

```python
mse_values = []
for c in c_values:
    mse = np.mean((Y - c)**2)
    mse_values.append(mse)
    
```

4.3

```python
best_index = np.argmin(mse_values)  
optimal_c = c_values[best_index]     
my_mean = np.mean(Y)
if optimal_c > my_mean:
    print("Optimal_c is bigger by:" , optimal_c - my_mean)
elif optimal_c < my_mean:
    print("My_mean is bigger by: ",my_mean - optimal_c)
elif optimal_c == my_mean:
    print("They are equal!")
```

4.4.Plot the MSE values against the different values of c.Mark the optimal 
c on the plot.

```python
plt.plot(c_values, mse_values, marker='o', label="MSE values")
plt.axvline(optimal_c, color='red', linestyle='--', label=f"Optimal c = {optimal_c:.2f}")
plt.scatter(optimal_c, mse_values[best_index], color='red', s=100, zorder=5)

plt.xlabel("c values")
plt.ylabel("MSE")
plt.title("MSE vs c")
plt.legend()
plt.show()
```
