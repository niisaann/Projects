```python
import numpy as np
predictors = ['zn','indus','chas','nox',"rm","age","dis","rad","tax","ptratio","lstat","medv"]
X_design = np.column_stack([Boston[p] for p in predictors])
X_design = sm.add_constant(X_design)

model = sm.OLS(y, X_design).fit()
print(model.summary())
```

(c) How do your results from (a) compare to your results from (b)? Create a plot displaying the univariate regression coefficients from (a) on the x-axis, and the multiple regression coefficients from (b) on the y-axis. That is, each predictor is displayed as a single point in the plot. Its coefficient in a simple linear regression model is shown on the x-axis, and its coefficient estimate in the multiple linear regression model is shown on the y-axis.

```python
models = [model2, model3, model4, model5, model6, model7, model8, model9, model10, model11, model12, model13]
betas_simple = []

for m in models:
    betas_simple.append(m.params[1])

betas_multi = model.params[1:]

import matplotlib.pyplot as plt

predictor_names = ["zn","indus","chas","nox","rm",
                   "age","dis","rad","tax","ptratio","lstat"]

plt.figure(figsize=(10,10))
plt.scatter(betas_simple, betas_multi)


for i, name in enumerate(predictor_names):
    plt.text(betas_simple[i], betas_multi[i], name, fontsize=10,
             ha='right', va='bottom')


lims = [min(betas_simple+betas_multi), max(betas_simple+betas_multi)]
plt.plot(lims, lims, '--', color='red')

plt.xlabel("Univariate coefficients (from part a)")
plt.ylabel("Multiple regression coefficients (from part b)")
plt.title("Comparison of coefficients: simple vs multiple regression")
plt.grid(alpha=0.3)
plt.show()

```

```python
plt.figure(figsize=(8,8))

colors = ["red" if b<0 else "blue" for b in betas_simple]

plt.scatter(betas_simple, betas_multi, s=80, alpha=0.8, c=colors, edgecolor="black")

for i, name in enumerate(predictor_names):
    plt.text(betas_simple[i]+0.05, betas_multi[i]+0.05, name, fontsize=10)

lims = [-5,10]
plt.plot(lims, lims, '--', color='grey')

plt.xlim(lims)
plt.ylim(lims)

plt.xlabel("Univariate beta (simple regression)", fontsize=12)
plt.ylabel("Multiple beta (full model)", fontsize=12)
plt.title("Comparison of coefficients", fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

We compare the univariate regression coefficients with the corresponding coefficients from the multiple regression model by plotting them against each other, one point per predictor. The x-axis shows the beta estimates obtained from the simple regressions in part (a), and the y-axis shows the beta estimates from the full multiple regression in part (b).

Several points lie far from the 45-degree line, indicating that their estimated effects change substantially once the other predictors are included in the model. This suggests the presence of multicollinearity and overlapping explanatory power among predictors. Predictors whose points lie closer to the diagonal have similar coefficients in both models, meaning that controlling for other variables does not strongly affect their estimated effect on the response.

(d)Is there evidence of non-linear association between any of the predictors and the response? To answer this question, for each predictor 
X,fit a model of the form smth..

```python
non_linear = []   

for col in Boston.columns:
    if col == "crim":
        continue  
    x = Boston[col].values
    X_poly = np.column_stack([x, x**2, x**3])
    X_poly = sm.add_constant(X_poly)

    model1 = sm.OLS(y, X_poly).fit()
    p2 = model1.pvalues[2]
    p3 = model1.pvalues[3]

    if p2 < 0.05 or p3 < 0.05:
        non_linear.append((col, p2, p3))

print("Predictors with evidence of non-linear relationship (p<0.05):")
for name, p2, p3 in non_linear:
    print(f"- {name}:  p(X^2)={p2:.4f},  p(X^3)={p3:.4f}")
```
